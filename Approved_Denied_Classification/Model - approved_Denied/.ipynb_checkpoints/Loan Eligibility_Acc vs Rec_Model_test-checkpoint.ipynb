# Model building

```python
# identify Dependent and independent variable
y = model_data.Loan_Status
X = model_data.drop('Loan_Status',axis=1)

# might decide to drop state
```

```python
X = pd.get_dummies(X)
```

```python
X
```

```python
# Split to test/train split and scale
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, random_state=42)
```

### Stochastic Gradient Descent (SGD) classifier

This classifier has the advantage of being capable of handling very large datasets efficiently. This is in part because SGD deals with training instances independently.

It is also very easy on memory and can provide an answer quickly. 

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)

sgd_clf.fit(X_train, y_train)
```

```python
# for testing purposes - Lets see how good/bad the model is at predicting 
y_train_acc = (y_train == 1)
y_train_rej = (y_train == 0)

# for testing purposes - later
y_test_acc = (y_test == 1)
y_test_rej = (y_test == 0)
```

```python
# cross validate to see performance 
# lets see how well the model predicts approvals
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_train, y_train_acc, cv=3, scoring="accuracy")
```

about 86% accuracy. Seems like a good score. But lets look further

#### Confusion Matrix

```python
# fit a confusion matrix for use see how well it predicts 1 of the targets (approvals)
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_acc, cv=3)
```

```python
# create a confustion matrix after sgd prediction

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_train_acc, y_train_pred)
```

```python
cm_display = ConfusionMatrixDisplay(cm).plot()
```

[**TRUE NEGATIVE** (*416,393*)     ,**FALSE POSITIVE**  (20,179)]

[**FALSE NEGATIVE** (81,708)   ,**TRUE POSITIVE**  (*257,397*)]

```python
# if the perfect prediction existed - IT would look like the below.
y_train_perfect_predictions = y_train_acc
perfect_cm = confusion_matrix(y_train_acc, y_train_perfect_predictions)
perfect_cm_display = ConfusionMatrixDisplay(perfect_cm).plot()
```

#### Precision and Recall

```python
from sklearn.metrics import precision_score, recall_score
print("Precision score: ",precision_score(y_train_acc, y_train_pred),"---> What percent of your predictions were correct?")
print("Recall: ",recall_score(y_train_acc, y_train_pred), "---> What percent of the positive cases did you catch? ")
```

combine precision and recall into a single metric called the F1 score, in particular if you need a simple way to compare two classifiers. The F1 score is the harmonic mean of precision and recall. Whereas the regular mean treats all values equally, the harmonic mean gives much more weight to low values. As a result, the classifier will only get a high F1 score if both recall and precision are high.

```python
from sklearn.metrics import f1_score
f1_score(y_train_acc, y_train_pred)
```

### Random Forest Classifier

Letâ€™s train a RandomForestClassifier and compare its ROC curve and ROC AUC score to the SGDClassifier.

The receiver operating characteristic (ROC) curve is another common tool used with
binary classifiers. It is very similar to the precision/recall curve, but instead of plotting
precision versus recall, the ROC curve plots the true positive rate (another name
for recall) against the false positive rate.

Because there are some trade-offs between precision/recall scores, ROC AUC might be the best option to evaluate Both the SGD model and the RFC model.

```python
# create and fit an RFC model
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(random_state=42)

y_probas_forest = cross_val_predict(forest_clf, X_train, y_train_acc, cv=3, method="predict_proba")
```

```python
# to plot a ROC curve to see score for RFC vs SGD
from sklearn.metrics import plot_roc_curve

rfc = RandomForestClassifier(n_estimators=10, random_state=42)
rfc.fit(X_train, y_train)

ax = plt.gca()
rfc_disp = plot_roc_curve(rfc, X_test, y_test, ax=ax, alpha=0.8)
sgd_clf_disp = plot_roc_curve(sgd_clf, X_test, y_test, ax=ax, alpha=0.8)
plt.show()
```

One way to compare classifiers is to measure the area under the curve (AUC). A perfect classifier will have a ROC AUC equal to 1

So here we see that RandomForest Classifier is a better model

## What this means


#### Precision - What percent of your predictions were correct?  
Precision is the ability of a classifier not to label an instance positive that is actually negative. For each class it is defined as the ratio of true positives to the sum of true and false positives. TP/(TP + FP)

#### Recall = What percent of the positive cases did you catch? 
Recall is the ability of a classifier to find all positive instances. For each class it is defined as the ratio of true positives to the sum of true positives and false negatives. TP/(TP+FN)

#### F1 score â€“ What percent of positive predictions were correct? 
The F1 score is a weighted harmonic mean of precision and recall such that the best score is 1.0 and the worst is 0.0. Generally speaking, F1 scores are lower than accuracy measures as they embed precision and recall into their computation. As a rule of thumb, the weighted average of F1 should be used to compare classifier models, not global accuracy.
